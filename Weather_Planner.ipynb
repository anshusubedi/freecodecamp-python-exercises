# Build a Travel Weather Planner

## Using conditional statements to determine whether commuting is possible based on the weather, the distance to travel, and the availability of a vehicle.

This program uses conditional statements to decide whether commuting is possible based on distance, weather, and available transportation. It first checks if the distance is a falsy value, returning False if so. Then, it evaluates distance ranges from shortest to longest: for short distances, commuting is only allowed if it is not raining; for medium distances, a bike is required and it must not be raining; and for long distances, either a car or a ride-share app must be available. By using logical operators like and, or, and not, the program ensures each condition is properly combined to produce an accurate result.

```python
# Create the variables
distance_mi = 3
is_raining = False
has_bike = True
has_car = False
has_ride_share_app = False
```

```python
#If distance_mi is a falsy value: print False.
# If the distance is less than or equal to 1 mile, print True only if it is not raining. Otherwise, you should print False.
# If the distance is greater than 1 mile and less than or equal to 6 miles, print True only if the person has a bike and it is not raining. 
#...Otherwise print False.
# If the distance is greater than 6 miles:print True if the person has a car or has a ride-share app, otherwise print False.

if not distance_mi:
    print(False)

elif distance_mi <= 1:
    if not is_raining:
        print(True)
    else:
        print(False)

elif distance_mi <= 6:
    if has_bike and not is_raining:
        print(True)
    else:
        print(False)

else:
    if has_car or has_ride_share_app:
        print(True)
    else:
        print(False)
```
