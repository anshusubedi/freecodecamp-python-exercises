# Jupyter Notebook Project: Apply Discount Function

## Apply Discount Calculator
This project implements a function that calculates the final price of an item after applying a percentage discount.
It includes:
- Input validation
- Discount calculation
- Example test cases

## Objective 
Create a function called `apply_discount(price, discount)` that:

- Validates inputs
- Applies a percentage discount
- Returns the final price

## Function Implementation

```python

def apply_discount(price, discount):
    # Validate price
    if not isinstance(price, (int, float)):
        return "The price should be a number"
    
    # Validate discount
    if not isinstance(discount, (int, float)):
        return "The discount should be a number"
    
    # Validate price value
    if price <= 0:
        return "The price should be greater than 0"
    
    # Validate discount range
    if discount < 0 or discount > 100:
        return "The discount should be between 0 and 100"
    
    # Calculate final price
    discount_amount = price * (discount / 100)
    final_price = price - discount_amount
    
    return final_price
```

## Test Cases

```python
# Valid cases
print(apply_discount(100, 20))   # Expected: 80
print(apply_discount(200, 50))   # Expected: 100
print(apply_discount(50, 0))     # Expected: 50
print(apply_discount(100, 100))  # Expected: 0
print(apply_discount(74.5, 20))  # Expected: 59.6

# Invalid cases
print(apply_discount("100", 20))   # Invalid price
print(apply_discount(100, "20"))   # Invalid discount
print(apply_discount(-10, 20))     # Invalid price value
print(apply_discount(100, 120))    # Invalid discount range
```

## Explanation

1. The function checks if inputs are numbers.
2. It ensures:
   - Price > 0
   - Discount is between 0 and 100
3. It calculates the discount:
   
   discount = price × (discount / 100)

4. It subtracts the discount from the original price.
